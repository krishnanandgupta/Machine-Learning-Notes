```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Create dataframe

df=pd.DataFrame({'A':[1,2,np.nan],'B':[5,np.nan,np.nan]})
df['States']="Ca NV AZ".split()
df.set_index("States",inplace=True)
print(df)
```

```python
#Remove row having na values 

df.dropna()
```

```python
#remove column having na 

df.dropna(axis=1)
```

```python
#Thresh : Requires that many Na values 

df.dropna(axis=0,thresh=1)
```

```python
#Fill na values

df.fillna(value="Krishna")
```

```python
#Fill na values with mean of column

df.fillna(value=df["A"].mean())
```
